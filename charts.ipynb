```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
file_path = "/content/drive/MyDrive/output_final.xlsx"
```

```python
data = pd.read_excel(file_path)
```

```python
#Pie Chart: Top 10 Zones by Entering Trips This chart can show the top 10 zones with the highest number of entering trips.
top_10_entering = data.nlargest(10, 'Entering')
plt.pie(top_10_entering['Entering'], labels=top_10_entering['Place'], autopct='%1.1f%%')
plt.title('Top 10 Zones by Entering Trips')
plt.show()
```

```python
#Histogram: Distribution of Entering and Leaving Trips This chart can visualize the distribution of entering and leaving trips across all zones.
data['Entering'].plot(kind='hist', bins=50)
data['Leaving'].plot(kind='hist', bins=50)
plt.title('Distribution of Entering and Leaving Trips')
plt.xlabel('Trips')
plt.ylabel('Frequency')
plt.show()
```

```python
data.plot(kind='bar', x='Place', y=['Entering', 'Leaving'], stacked=True)
plt.title('Entering and Leaving Trips by Zone')
plt.xlabel('Zone')
plt.ylabel('Trips')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.heatmap(data.pivot(index='Leaving', columns='Entering', values='annual_total_trips'), annot=True, cmap='coolwarm')
plt.title('Zone-to-Zone Trip Matrix')
plt.xlabel('Destination Zone')
plt.ylabel('Origin Zone')
plt.show()
```
